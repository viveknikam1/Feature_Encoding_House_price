```python
def create_dummies(df):
    i=0
    for column in df.columns:
        df_temp = pd.get_dummies(df[column],drop_first=True)
        if i==0:
            df_merged = df_temp
        else:
            df_merged = pd.concat([df_merged,df_temp],axis=1)
        i+=1
    return df_merged
```
